```
import torch
import torch.nn as nn
```

```
bceloss = nn.BCELoss(reduction='sum')

y1 = torch.tensor([0.99, 0.01])
y2 = torch.tensor([0.5, 0.5])
y3 = torch.tensor([0.99, 0.45])
y4 = torch.tensor([0.65, 0.65])

y_t = torch.tensor([1, 1], dtype=torch.float32)
```

```
final = list(map(lambda a: bceloss(a, y_t), [y1, y2, y3, y4]))
```

```
zipka = dict(zip(final, [y1, y2, y3, y4]))
```
